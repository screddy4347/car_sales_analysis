```python
plt.figure(figsize=(10,5))
df['brand'].value_counts().plot(kind='bar')
plt.title("Number of Cars by Brand")
plt.xlabel("Brand")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['selling_price'], kde=True)
plt.title("Selling Price Distribution")
plt.xlabel("Selling Price")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="fuel")
plt.title("Fuel Type Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,6))
df['fuel'].value_counts().plot(kind='pie', autopct="%1.1f%%")
plt.title("Fuel Type Distribution")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='km_driven', y='selling_price')
plt.title("KM Driven vs Selling Price")
plt.xlabel("KM Driven")
plt.ylabel("Selling Price")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["km_driven"], bins=30)
plt.title("KM Driven Distribution")
plt.xlabel("KM Driven")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="km_driven", y="selling_price", hue="fuel")
plt.title("Car Price vs KM Driven (Fuel Type)")
plt.show()
```

```python
sns.pairplot(df, hue="fuel")
plt.show()

```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='owner')
plt.title("Ownership Category Distribution")
plt.xlabel("Owner Type")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x="owner", y="selling_price")
plt.title("Selling Price by Owner Type")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
price_array = df['selling_price'].values

print("\n===== NUMPY PRICE ANALYSIS =====")
print("Max Price  :", np.max(price_array))
print("Min Price  :", np.min(price_array))
print("Mean Price :", np.mean(price_array))
```
